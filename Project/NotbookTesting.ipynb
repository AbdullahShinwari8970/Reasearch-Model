```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the updated CSV file
data = pd.read_csv('hyperparameter_errors.csv')

# Pivot the data to reshape it for the heatmap
pivot_data = data.pivot(index='Hidden Units', columns='Learning Rate', values='Error')

# Create a heatmap with a custom color palette
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, annot=True, cmap="coolwarm", fmt=".3f", cbar_kws={'label': 'Error'}, linewidths=0.5, linecolor='white')

# Adjusting the appearance for a cleaner result
plt.title('Hyperparameter Tuning Heatmap (XOR) - 10,000 Epochs')
plt.xlabel('Learning Rate')
plt.ylabel('Hidden Units')

# Adjust grid and layout
plt.grid(False)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the updated CSV file
data = pd.read_csv('hyperparameter_errors.csv')

# Pivot the data to reshape it for the heatmap
pivot_data = data.pivot(index='Hidden Units', columns='Learning Rate', values='Error')

# Normalize the error values (optional, but helpful for better visualization)
pivot_data_normalized = (pivot_data - pivot_data.min().min()) / (pivot_data.max().max() - pivot_data.min().min())

# Create a heatmap with a custom color palette
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data_normalized, annot=True, cmap="coolwarm", fmt=".3f", cbar_kws={'label': 'Normalized Error'}, linewidths=0.5, linecolor='white')

# Adjusting the appearance for a cleaner result
plt.title('Hyperparameter Tuning Heatmap (XOR) - Normalized Errors')
plt.xlabel('Learning Rate')
plt.ylabel('Hidden Units')

# Adjust grid and layout
plt.grid(False)
plt.tight_layout()

# Show the plot
plt.show()
```
